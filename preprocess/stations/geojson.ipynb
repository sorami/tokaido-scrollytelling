```python
import pandas as pd
import geopandas
```

```python
df = pd.read_csv("stations.csv")
df.head().T
```

```python
gdf = geopandas.GeoDataFrame(
    df["no"],
    geometry=geopandas.points_from_xy(
        df.longitude,
        df.latitude
    )
)
gdf.head().T
```

```python
gdf.to_file("../../public/stations/points.geojson")
```
